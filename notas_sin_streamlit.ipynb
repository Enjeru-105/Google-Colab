```python
class Nota:
  def __init__(self, titulo, contenido):
    self.titulo = titulo
    self.contenido = contenido

  def mostrar(self):
    print(f"Titulo: {self.titulo}")
    print(f"Contenido: {self.contenido}")

  def modificar_t(self, titulo):
    self.titulo = titulo

  def modificar_c(self, contenido):
    self.contenido = contenido

class NotaTexto(Nota):
  def __init__(self, titulo, contenido):
    super().__init__(titulo, contenido)

  def mostrar(self):
    super().mostrar()

  def modificar_t(self, titulo):
    super().modificar_t(titulo)

  def modificar_c(self, contenido):
    super().modificar_c(contenido)

class NotaLista(Nota):
  def __init__(self, titulo, contenido=None):
    super().__init__(titulo, contenido)
    self.contenido = contenido if contenido is not None else []

  def mostrar(self):
    print(f"Titulo: {self.titulo}")
    print("Contenido (Lista):")
    for i, item in enumerate(self.contenido, 1):
      print(f"{i}. {item}")

  def modificar_t(self, titulo):
    super().modificar_t(titulo)

  def modificar_c(self, contenido):
    super().modificar_c(contenido)

class NotaImagen(Nota):
  def __init__(self, titulo, contenido):
    super().__init__(titulo, contenido)
    #Contenido es una URL

  def mostrar(self):
    super().mostrar()

  def modificar_t(self, titulo):
    super().modificar_t(titulo)

  def modificar_c(self, contenido):
    super().modificar_c(contenido)

class BlocDeNotas:
  def __init__(self):
    self.notas = []

  def agregar_nota(self, nota):
    self.notas.append(nota)

  def existe_nota(self, titulo):
    return any(nota.titulo == titulo for nota in self.notas)

#---------main---------#
NotaText = NotaTexto("","")
NotaList = NotaLista("",[])
NotaImage = NotaImagen("","")
BlocDeNotas = BlocDeNotas()

while True:
  if len(BlocDeNotas.notas) == 0:
    print("\nNo hay notas guardadas\n")
  opc= input("\nQue desea hacer?\n1. Agregar nota\n2. Mostrar Todas las Notas\n3. Editar Nota\n4. Eliminar Notas\n0. Salir\n\n")
  match opc:
    case "1":
      opc2 = input("Que tipo de nota desea agregar?\n1. Nota de Texto\n2. Nota de Lista\n3. Nota de Imagen\n\n")
      match opc2:
        case "1":
          print("\nNota de Texto\n")
          while True:
            titulo = input("\nIngrese el titulo de la nota:\n")
            if BlocDeNotas.existe_nota(titulo):
              print("\nYa existe una nota con ese titulo, use otro!\n")
            else:
              break
          contenido = input("Ingrese el contenido de la nota (Sin saltos de línea)\n")
          NotaText = NotaTexto(titulo, contenido)
          BlocDeNotas.agregar_nota(NotaText)

        case "2":
          print("\nNota de Lista\n")
          while True:
            titulo = input("\nIngrese el titulo de la nota:\n")
            if BlocDeNotas.existe_nota(titulo):
              print("\nYa existe una nota con ese titulo, use otro!\n")
            else:
              break
          contenido = []
          cont = 1
          print("Si desea dejar de agregar elementos a la lista, introduzca un 0")
          while True:
            contenido.append(input(f"\nIngrese el {cont}º elemento de la lista:"))
            if contenido[-1] == "0":
              contenido.pop()
              break
            else:
              cont += 1
          NotaList = NotaLista(titulo, contenido)
          BlocDeNotas.agregar_nota(NotaList)

        case "3":
          print("\nNota de Imagen\n")
          while True:
            titulo = input("\nIngrese el titulo de la nota:\n")
            if BlocDeNotas.existe_nota(titulo):
              print("\nYa existe una nota con ese titulo, use otro!\n")
            else:
              break
          contenido = input("Ingrese el URL de la imagen a guardar:\n")
          NotaImage = NotaImagen(titulo, contenido)
          BlocDeNotas.agregar_nota(NotaImage)

        case _:
          print("\nOpcion no valida\n\n")

    case "2":
      print("\nNotas Guardadas:\n\n")
      cont = 1
      for nota in BlocDeNotas.notas:
        print(f"Nota {cont}:")
        cont += 1
        nota.mostrar()
        print("\n")

    case "3":
      buscar = input("\nIngrese el titulo de la nota a editar:\n")
      nota_encontrada = None
      for nota in BlocDeNotas.notas:
        if nota.titulo == buscar:
          nota_encontrada = nota
          break
      if nota_encontrada:
        print("\nNota encontrada:\n")
        nota_encontrada.mostrar()

        while True:
          nuevo_titulo = input("\nIngrese el nuevo título (deje vacío para mantener el actual):\n")
          if nuevo_titulo:
            if BlocDeNotas.existe_nota(nuevo_titulo):
              print("\nYa existe una nota con ese título, use otro!\n")
            else:
              nota_encontrada.modificar_t(nuevo_titulo)
              break
          else:
            break

        if isinstance(nota_encontrada, NotaTexto):
          nuevo_contenido = input("Ingrese el nuevo contenido de la nota (deje vacío para mantener el actual):\n")
          if nuevo_contenido:
            nota_encontrada.modificar_c(nuevo_contenido)

        elif isinstance(nota_encontrada, NotaLista):
          print("Contenido actual de la lista:")
          for i, item in enumerate(nota_encontrada.contenido, 1):
            print(f"{i}. {item}")
          nueva_lista = []
          print("Ingrese los nuevos elementos de la lista (ingrese '0' al inicio para mantener el actual o al último para finalizar):")
          while True:
            item = input(f"\nNuevo elemento {len(nueva_lista) + 1}: ")
            if item == "0":
              break
            elif item:
              nueva_lista.append(item)
          if nueva_lista:
            nota_encontrada.modificar_c(nueva_lista)

        elif isinstance(nota_encontrada, NotaImagen):
          nueva_url = input("Ingrese la nueva URL de la imagen (deje vacío para mantener la actual):\n")
          if nueva_url:
            nota_encontrada.modificar_c(nueva_url)
        print("\nNota actualizada correctamente!\n")

      else:
        print("\nNo se encontraron notas con ese título\n")

    case "4":
      opc3 = input("Estás seguro de eliminar todas las notas guardadas?\n1. Si\n2. No\n\n")
      match opc3:
        case "1":
          BlocDeNotas.notas = []
          print("\nNotas eliminadas\n")

        case "2":
          print("\nNo se eliminaron las notas\n")

        case _:
          print("\nOpcion no valida\n\n")

    case "0":
      break

    case _:
      print("\nOpcion no valida\n\n")
```
